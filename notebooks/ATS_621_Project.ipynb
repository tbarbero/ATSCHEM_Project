```python
import icartt
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER
import matplotlib.ticker as mticker
from matplotlib import rc
import matplotlib.patches as mpatches
import matplotlib.path as mpath
import matplotlib as mpl
import matplotlib.gridspec as gs
import matplotlib
import warnings
from fig_params import *
fig_params()
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.simplefilter("ignore",UserWarning)
warnings.simplefilter("ignore",RuntimeWarning)

import matplotlib.patheffects as PathEffects
from owslib.wmts import WebMapTileService

def setup_m(ax):
    ax.set_global()
    ax.add_feature(cfeature.COASTLINE)
    # https://github.com/SciTools/cartopy/issues/1366
#     ax.add_feature(cfeature.STATES, zorder=1, linewidth=1, edgecolor='k')
    ax.add_feature(cfeature.STATES.with_scale('110m'), zorder=3, linewidth=0.6, edgecolor='k')
    t1 = ax.gridlines(crs=ccrs.PlateCarree(),draw_labels=True,alpha=0.2)
#     t1.top_labels = False
#     t1.right_labels = False
    t1.xlabels_top = False
    t1.ylabels_right = False
    t1.xformatter = LONGITUDE_FORMATTER
    t1.yformatter = LATITUDE_FORMATTER
    t1.ylocator = mticker.FixedLocator(np.arange(-90,90,10))
    t1.xlocator = mticker.FixedLocator(np.arange(-180,180,20))
    return ax
```

```python
def get_dates(filelist):
    
    # sort files by date
    filelist = sorted(filelist)
    
    # parse yyyymmdd and put into list
    dates = [filelist[i][13:21] for i,file in enumerate(filelist)]
    return dates

def create_dataarray(filelist='',nmax=500,varname=''):
    # create a xr.DataArray composed of all the data
    data = np.empty([len(filelist),nmax]);data[:]=np.NaN;
    da = xr.DataArray(
        data=data,
        dims=["date","time"],
        coords=dict(
            date=dates,
#             time=np.arange(0,nmax,1)
                    )
                )
    da.name=varname
    return da
```

```python
rootdir = "/Users/tyler/Desktop/621/FinalProject/data/"

# get dates from flight legs
dates = get_dates(filelist)

# create empty dataset
ds = xr.Dataset(data_vars=None,
                    coords=dict(date=dates),
                    attrs=None)

# get maximum array size length in order to set max length of DataArray
tmp = []
for filename in filelist:
    ict = icartt.Dataset(rootdir+filename)
    tmp.append(int(np.shape(ict.data['G_LONG'])[0]))
nmax = np.max(tmp)

for var in varlist:
    # assign memory for each variable [len(filelist),n]
    data = np.empty([len(filelist),nmax]);data[:]=np.NaN
    
    for i,file in enumerate(sorted(filelist)):
        ict = icartt.Dataset(rootdir+file)
        n = len(ict.data['G_LONG'])
        data[i,:n]=ict.data[var]
    
    # get attributes of each var
    units = ict.variables[var].units
#     longname = ict.variables[var].longname
    
    # now save  variable in dataset 
    ds[var] = (["date","time"],data)
    ds[var].attrs = {"units":units}
```
